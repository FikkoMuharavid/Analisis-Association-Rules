# **Asosiation Rule**

## Import Library dan Dataset

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Loading the dataset

data = pd.read_csv('/content/Bakery.csv')
```

```python
data.head()
```

```python
# The of the data

data.shape
```

## Preprocess Data

Penanganan Missing Value

```python
print("Data yang Hilang:\n",data.isnull().values.any())
seed_clean = data.dropna()
```

```python
# As the previous cell told us that there are some null values. So, let's find them!

data.isnull().sum()
```

Menyesuaikan Tipe Data

```python
# Stripping extra spaces in the description
data['Items'] = data['Items'].str.strip()

# Dropping the rows without any invoice number
data.dropna(axis = 0, subset =['TransactionNo'], inplace = True)
data['TransactionNo'] = data['TransactionNo'].astype('str')
```

```python
data['Quantity'] = 1
data.head()
```

Memisah Atribut yang Akan Digunakan

```python
weekend = (data[data['DayType'] =="Weekend"]
          .groupby(['TransactionNo', 'Items'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TransactionNo'))
```

Encode Dataset

```python
# Defining the hot encoding function to make the data suitable

def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1

# Applying one hot encoding

shop_encoded = weekend.applymap(hot_encode)
weekend = shop_encoded
```

```python
weekend.head()
```

## Melakukan Asosiation Rule

```python
# Building the model
frq_items = apriori(weekend, min_support = 0.7, use_colnames = True)

# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 0.7)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
```
